```python
# 2. í‰ê·  êµ¬í•˜ê¸°

# ì‚¬ìš©ìë¡œë¶€í„° ìˆ«ì ì—¬ëŸ¬ ê°œë¥¼ ì…ë ¥ë°›ê³ , ì…ë ¥í•œ ìˆ«ìë“¤ì˜ í‰ê· ì„ êµ¬í•˜ëŠ” í”„ë¡œê·¸ë¨ì´ì•¼.
# ì˜ˆì‹œ ì…ë ¥: 5, 10, 15
# ì˜ˆì‹œ ì¶œë ¥: í‰ê· ì€ 10ì…ë‹ˆë‹¤.

# 1. ìˆ«ì ì…ë ¥, ì—¬ëŸ¬ê°œ  
# 2. ì •ìˆ˜ë¡œ ë³€í™˜
# 3. ê²°ê³¼ ì¶œë ¥
# 4. f ìŠ¤íŠ¸ë§ ìœ¼ë¡œ í‰ê· ì€ {/} ì…ë‹ˆë‹¤.

user_input = input("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”. (ì‰¼í‘œë¡œ êµ¬ë¶„): ")

try:
    # ì‰¼í‘œë¡œ ë‚˜ëˆ ì„œ ìˆ«ìë¡œ ë³€í™˜
    numbers = [float(x.strip()) for x in user_input.split(",")]

    # ë¹ˆ ì…ë ¥ ì²˜ë¦¬
    if len(numbers) == 0:
        print("ìˆ«ìë¥¼ ì…ë ¥í•´ì•¼ í•©ë‹ˆë‹¤!")
        exit()

    # í‰ê·  ê³„ì‚°
    average = sum(numbers) / len(numbers)
    print(f"ì…ë ¥í•œ ìˆ«ìì˜ í‰ê· ì€ {average}ì…ë‹ˆë‹¤.")

except ValueError:
    print("ìˆ«ìë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”!")
    exit()
```

```python
# ë¬¸ì œ 1: ì†Œìˆ˜ íŒë³„í•˜ê¸°

# ì‚¬ìš©ìë¡œë¶€í„° ìˆ«ìë¥¼ ì…ë ¥ë°›ì•„ í•´ë‹¹ ìˆ«ìê°€ ì†Œìˆ˜ì¸ì§€ ì•„ë‹Œì§€ íŒë³„í•˜ëŠ” í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•˜ì„¸ìš”.
# ì†Œìˆ˜(Prime Number)ëŠ” 1ê³¼ ìê¸° ìì‹ ìœ¼ë¡œë§Œ ë‚˜ëˆ„ì–´ì§€ëŠ” ìˆ«ìì…ë‹ˆë‹¤.
# ì˜ˆ: 2, 3, 5, 7, 11 ë“±.

# 1. ìˆ«ì ì…ë ¥ input
# 2. ì†Œìˆ˜ êµ¬ë³„ % ë‚˜ëˆ„ê¸°ë¡œ??
# 3. ìˆ«ì ì¶œë ¥

num = int(input("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš” : "))

# ì†Œìˆ˜ì¸ì§€ íŒë³„í•˜ëŠ” ë³€ìˆ˜
is_prime = True

if num < 2: # 2ë³´ë‹¤ ì‘ì€ ìˆ«ìëŠ” ì†Œìˆ˜ê°€ ì•„ë‹˜
    is_prime = False
else:
    for i in range(2, int(num ** 0.5) + 1): # 2ë¶€í„° numì˜ ì œê³±ê·¼ê¹Œì§€ í™•ì¸
        if num % i == 0:
            is_prime = False
            break

# ê²°ê³¼ ì¶œë ¥
if is_prime:
    print(f"{num}ì€ ì†Œìˆ˜ì…ë‹ˆë‹¤.")
else:
    print(f"{num}ì€ ì†Œìˆ˜ê°€ ì•„ë‹™ë‹ˆë‹¤.")
```

```python
# ë¬¸ì œ 4: í™€ìˆ˜ì˜ í•© êµ¬í•˜ê¸°

# 1ë¶€í„° ì‚¬ìš©ìê°€ ì…ë ¥í•œ ìˆ«ìê¹Œì§€ì˜ ëª¨ë“  í™€ìˆ˜ì˜ í•©ì„ êµ¬í•˜ëŠ” í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•˜ì„¸ìš”.
# ì…ë ¥ ì˜ˆì‹œ: 10
# ì¶œë ¥ ì˜ˆì‹œ: 25 (1 + 3 + 5 + 7 + 9)

# 1. ì…ë ¥ ë°›ì„ ê³³
# 2. í™€ìˆ˜ë§Œ ê³¨ë¼ë‚´ì„œ ë”í•˜ê¸°
# 3. ì¶œë ¥ ë°˜í™˜

num = int(input("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš” : "))

sum_of_odds = 0     # í™€ìˆ˜ì˜ í•©ì„ ì €ì¥í•  ë³€ìˆ˜

for i in range(1, num + 1):     # 1ë¶€í„° ì…ë ¥ê°’ê¹Œì§€ ë°˜ë³µ
    if i % 2 != 0:  # í™€ìˆ˜ë¼ë©´
        sum_of_odds += i    # í•©ì— ë”í•˜ê¸°

print(f"1ë¶€í„° {num}ê¹Œì§€ì˜ í™€ìˆ˜ í•©ì€ {sum_of_odds}ì…ë‹ˆë‹¤.")
```

```python
# ğŸ”¹ ë¬¸ì œ 1: ì‚¬ê°í˜• í´ë˜ìŠ¤ ë§Œë“¤ê¸°
# 	ì‚¬ê°í˜•ì˜ **ê°€ë¡œ(width)**ì™€ **ì„¸ë¡œ(height)**ë¥¼ ì…ë ¥ë°›ì•„ **ë„“ì´(area)**ë¥¼ ê³„ì‚°í•˜ëŠ” í´ë˜ìŠ¤ë¥¼ ë§Œë“¤ì–´ë¼.

# ìš”êµ¬ì‚¬í•­:
# 	1.	í´ë˜ìŠ¤ ì´ë¦„ì€ Rectangle.
# 	2.	ìƒì„±ìë¥¼ ì‚¬ìš©í•´ ê°€ë¡œì™€ ì„¸ë¡œë¥¼ ì´ˆê¸°í™”.
# 	3.	ë„“ì´ë¥¼ ë°˜í™˜í•˜ëŠ” ë©”ì„œë“œ get_area()ë¥¼ ì‘ì„±.

class Rectangle:
    def __init__(self, width, height):
        self.width = width   # ê°€ë¡œë¥¼ ì†ì„±ìœ¼ë¡œ ì €ì¥
        self.height = height    # ì„¸ë¡œë¥¼ ì†ì„±ìœ¼ë¡œ ì €ì¥

    def get_area(self):
        return self.width * self.height     # ë„“ì´ ê³„ì‚°

# ì‚¬ìš©ì ì…ë ¥
num1 = int(input("ê°€ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš” : "))
num2 = int(input("ì„¸ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš” : "))

rect = Rectangle(num1, num2)
print(f"ì‚¬ê°í˜•ì˜ ë„“ì´ëŠ” {rect.get_area()}")
```
